```r
# ข้อมูล
data <- c(1.41, 0.26, 1.97, 0.33, 0.55, 0.77, 1.46, 1.18)
n <- length(data)

# การแจกแจงสะสม (CDF) ของข้อมูลตาม Uniform(0, 2)
cdf_theoretical <- data / 2

# การแจกแจงสะสมเชิงประจักษ์ (Empirical CDF)
cdf_empirical <- rank(data) / n

# คำนวณสถิติ D
D <- max(abs(cdf_empirical - cdf_theoretical))

# คำนวณค่า critical value ที่ระดับนัยสำคัญ 0.05
alpha <- 0.05
critical_value <- 1.36 / sqrt(n)

# แสดงค่า D และ critical value
print(paste("D: ", D))
print(paste("Critical value: ", critical_value))

# สรุปผลการทดสอบ
if (D > critical_value) {
  print("ปฏิเสธ H0: ข้อมูลไม่มาจากการแจกแจงแบบ Uniform(0,2)")
} else {
  print("ไม่ปฏิเสธ H0: ข้อมูลมาจากการแจกแจงแบบ Uniform(0,2)")
}
```

```r
# ข้อมูล
female <- c(66, 77, 57, 62, 68, 78, 73, 81, 84, 69, 62, 79)
male <- c(79, 71, 68, 71, 68, 86, 73, 58, 68, 74, 78)

# รวมข้อมูลทั้งหมดและหาค่าลำดับ
all_data <- c(female, male)
ranks <- rank(all_data)

# แยกอันดับของผู้หญิง
ranks_female <- ranks[1:length(female)]

# คำนวณ U สำหรับผู้หญิง
n1 <- length(female)
n2 <- length(male)
U_female <- n1 * n2 + (n1 * (n1 + 1)) / 2 - sum(ranks_female)

# คำนวณค่า U critical value ที่ระดับนัยสำคัญ 0.05
critical_value <- qnorm(0.05) * sqrt(n1 * n2 * (n1 + n2 + 1) / 12) + (n1 * n2) / 2

# แสดงค่า U และ critical value
print(paste("U_female: ", U_female))
print(paste("Critical value: ", critical_value))

# สรุปผลการทดสอบ
if (U_female < critical_value) {
  print("ปฏิเสธ H0: การเต้นของชีพจรของผู้หญิงน้อยกว่าผู้ชาย")
} else {
  print("ไม่ปฏิเสธ H0: การเต้นของชีพจรของผู้หญิงไม่แตกต่างจากผู้ชาย")
}
```

## ข้อที่ 6

```r
# ข้อมูล
florida <- c(3718, 4948, 9489, 8771, 13729, 6865, 7184)
nc <- c(4883, 5825, 2153, 3424, 2108)
texas <- c(8150, 4403, 3274, 2276, 3070, 3855)

# รวมข้อมูลและอันดับ
all_data <- c(florida, nc, texas)
ranks <- rank(all_data)

# ขนาดกลุ่ม
n_florida <- length(florida)
n_nc <- length(nc)
n_texas <- length(texas)

# หาผลรวมอันดับแต่ละกลุ่ม
R_florida <- sum(ranks[1:n_florida])
R_nc <- sum(ranks[(n_florida+1):(n_florida+n_nc)])
R_texas <- sum(ranks[(n_florida+n_nc+1):length(all_data)])

# คำนวณสถิติ H ของ Kruskal-Wallis
n <- length(all_data)
H <- (12 / (n * (n + 1))) * (R_florida^2 / n_florida + R_nc^2 / n_nc + R_texas^2 / n_texas) - 3 * (n + 1)

# ค่า critical value ที่ df = 2
critical_value <- qchisq(1 - 0.05, df = 2)

# แสดงค่า H และ critical value
print(paste("H: ", H))
print(paste("Critical value: ", critical_value))

# สรุปผลการทดสอบ
if (H > critical_value) {
  print("ปฏิเสธ H0: จำนวนธุรกิจขนาดเล็กในแต่ละรัฐแตกต่างกัน")
} else {
  print("ไม่ปฏิเสธ H0: จำนวนธุรกิจขนาดเล็กในแต่ละรัฐมีค่าเท่ากัน")
}
```

```r
# ข้อมูล
literacy <- c(21, 48, 51, 71, 86, 87, 94, 98, 99, 100)
fertility <- c(6.69, 2.73, 4.72, 4, 2.2, 1.73, 2.41, 1.28, 2.09, 1.25)

# Pearson correlation
pearson_result <- cor.test(literacy, fertility, method = "pearson")
spearman_result <- cor.test(literacy, fertility, method = "spearman")

# แสดงผลลัพธ์การทดสอบ Pearson และ Spearman
print(paste("Pearson correlation: ", pearson_result$estimate))
print(paste("Pearson p-value: ", pearson_result$p.value))

print(paste("Spearman correlation: ", spearman_result$estimate))
print(paste("Spearman p-value: ", spearman_result$p.value))

# สรุปผลลัพธ์ Pearson
if (pearson_result$p.value < 0.05) {
  print("ปฏิเสธ H0 (Pearson): มีความสัมพันธ์เชิงเส้นระหว่างอัตราการรู้หนังสือของเพศหญิงและอัตราการเจริญพันธุ์")
} else {
  print("ไม่ปฏิเสธ H0 (Pearson): ไม่มีความสัมพันธ์เชิงเส้นระหว่างอัตราการรู้หนังสือของเพศหญิงและอัตราการเจริญพันธุ์")
}

# สรุปผลลัพธ์ Spearman
if (spearman_result$p.value < 0.05) {
  print("ปฏิเสธ H0 (Spearman): มีความสัมพันธ์แบบลำดับระหว่างอัตราการรู้หนังสือของเพศหญิงและอัตราการเจริญพันธุ์")
} else {
  print("ไม่ปฏิเสธ H0 (Spearman): ไม่มีความสัมพันธ์แบบลำดับระหว่างอัตราการรู้หนังสือของเพศหญิงและอัตราการเจริญพันธุ์")
}
```

```r
# ข้อมูล
data <- matrix(c(1, 9, 11, 3), nrow = 2)

# คำนวณค่า Chi-square
observed <- data
expected <- chisq.test(observed)$expected

chi_square <- sum((observed - expected)^2 / expected)

# ค่า critical value ที่ df = 1
critical_value <- qchisq(1 - 0.05, df = 1)

# แสดงค่า chi-square และ critical value
print(paste("Chi-square: ", chi_square))
print(paste("Critical value: ", critical_value))

# สรุปผลการทดสอบ
if (chi_square > critical_value) {
  print("ปฏิเสธ H0: สัดส่วนของการทบทวนบทเรียนระหว่างผู้ชายและผู้หญิงไม่เท่ากัน")
} else {
  print("ไม่ปฏิเสธ H0: สัดส่วนของการทบทวนบทเรียนระหว่างผู้ชายและผู้หญิงเท่ากัน")
}
```

## ข้อที่ 9

```r
# ข้อมูลคะแนนที่ได้จากผู้เชี่ยวชาญแต่ละคน (J1 ถึง J10) สำหรับสินค้าแต่ละชิ้น (P1 ถึง P6)
data <- matrix(c(
  7, NA, 5, 7, NA, 3,
  8, NA, 6, NA, 6, 4,
  5, NA, 4, 7, NA, 6,
  9, 7, 5, NA, 9, NA,
  NA, 9, 6, 7, 5, NA,
  6, 4, NA, 5, 4, 8,
  7, 9, 3, NA, NA, 7,
  9, 3, NA, 5, 6, NA,
  5, 4, 5, NA, NA, 6,
  7, NA, 8, NA, 7, 9
), nrow = 10, byrow = TRUE)

# ตั้งชื่อแถวและคอลัมน์ให้กับข้อมูล
rownames(data) <- paste("J", 1:10, sep = "")
colnames(data) <- paste("P", 1:6, sep = "")

# จัดอันดับภายในแต่ละแถว (ผู้เชี่ยวชาญ) โดยจัดอันดับเฉพาะค่าที่ไม่ใช่ NA
ranked_data <- t(apply(data, 1, function(x) rank(x, na.last = "keep")))

# คำนวณค่าเฉลี่ยอันดับของแต่ละสินค้า
mean_ranks <- colMeans(ranked_data, na.rm = TRUE)

# คำนวณจำนวนผู้เชี่ยวชาญ (n) และจำนวนสินค้าที่ประเมิน (k)
n <- nrow(data)
k <- ncol(data)

# คำนวณสถิติ Friedman chi-squared
Q <- (12 * n) / (k * (k + 1)) * sum((mean_ranks - (k + 1) / 2)^2)

# คำนวณค่า p-value
p_value <- 1 - pchisq(Q, k - 1)

# แสดงผลลัพธ์
cat("Friedman chi-squared:", Q, "\n")
cat("p-value:", p_value, "\n")

# ตรวจสอบผลลัพธ์
if (p_value < 0.05) {
  cat("ปฏิเสธ H0: มีความแตกต่างอย่างมีนัยสำคัญระหว่างสินค้าทั้งหมด\n")
} else {
  cat("ไม่สามารถปฏิเสธ H0: ไม่มีความแตกต่างอย่างมีนัยสำคัญระหว่างสินค้าทั้งหมด\n")
}
```

```r
set.seed(5685)
n <- 15
data <- round(rnorm(n, mean=10, sd=3), 2)
cat("ชุดข้อมูล = ",data,")\n")
hist(data)
# Chi-squared test
breaks <- seq(min(data), max(data), length.out = 6)
observed <- hist(data, breaks = breaks, plot = FALSE)$counts
xbar <- mean(data)
sd <- sd(data)
expected_probs <- pnorm(breaks, mean = xbar, sd = sd)
expected_probs <- diff(expected_probs)
expected <- expected_probs * length(data)
chisq_test <- chisq.test(x = observed, p = expected / sum(expected))
# ตรวจสอบค่าคาดหวัง
all_expected_greater_than_5 <- all(chisq_test$expected > 5)
all_expected_greater_than_5
# แสดงผลการทดสอบ Chi-squared
cat("ผลการทดสอบ\n")
if (chisq_test$p.value < 0.05) {
  cat("การทดสอบแบบ Chi-squared : ข้อมูลมีการแจกแจงแบบไม่ normal ที่ p-value =", chisq_test$p.value, "\n")
} else {
  cat("การทดสอบแบบ Chi-squared : ข้อมูลมีการแจกแจงแบบ normal ที่ p-value =", chisq_test$p.value, "\n")
  }
# Kolmogorov-Smirnov test
ks_test <- ks.test(data, "pnorm", mean = mean(data), sd = sd(data))

# แสดงผลการทดสอบ Kolmogorov-Smirnov
if (ks_test$p.value < 0.05) {
  cat("การทดสอบด้วยวิธี Kolmogorov-Smirnov : ข้อมูลมีการแจกแจงแบบไม่ normal ที่ p-value =", ks_test$p.value, "\n")
} else {
  cat("การทดสอบด้วยวิธี Kolmogorov-Smirnov : ข้อมูลมีการแจกแจงแบบ normal ที่ p-value =", ks_test$p.value, "\n")
}
```

```r
# กำหนด student ID
set.seed(5685)
# สุ่มข้อมูล
data1 <- round(runif(25, min = 0, max = 100), 2)
data2 <- round(runif(25, min = 0, max = 100), 2)
data1
data2
# การทดสอบสมมติฐาน
# 1. รวมข้อมูลจากสองกลุ่มเข้าด้วยกัน
combined_data <- c(data1, data2)
# 2. จัดเรียงข้อมูลทั้งหมดจากน้อยไปมาก
sorted_data <- sort(combined_data)
# 3. กำหนดตำแหน่ง (rank) สำหรับข้อมูลทั้งสองกลุ่ม
rank_data1 <- rank(c(data1, rep(NA, length(data2))), na.last = "keep")
rank_data2 <- rank(c(rep(NA, length(data1)), data2), na.last = "keep")
# 4. คำนวณค่า rank sum ของกลุ่มที่หนึ่ง (R1)
R1 <- sum(rank_data1, na.rm = TRUE)
# 5. คำนวณสถิติทดสอบ (test statistic)
n1 <- length(data1)
n2 <- length(data2)
U <- R1 - (n1 * (n1 + 1)) / 2
# 6. คำนวณค่า Var(U) และสถิติ Z
mean_U <- n1 * n2 / 2
var_U <- (n1 * n2 * (n1 + n2 + 1)) / 12
Z <- (U - mean_U) / sqrt(var_U)
# 7. ทดสอบสมมติฐานด้วยการแจกแจง Chi-squared
chi_squared <- Z^2
p_value <- 1 - pchisq(chi_squared, df = 1)
# 8. แสดงผลลัพธ์
cat("ค่า U =", U, "\n")
cat("ค่า Z =", Z, "\n")
cat("ค่า Chi-squared =", chi_squared, "\n")
cat("ค่า p-value =", p_value, "\n")
```

# Mann-Whitney Test

ทดสอบสมมติฐานว่ามัธยฐานของประชากร 2 กลุ่มเท่ากันหรือไม่ด้วยการทดสอบ Mann-Whitneyโดยสุ่มค่าได้ดังนี้

```r
set.seed(5685)
data1 <- round(runif(8, min = 0, max = 100), 0)
data2 <- round(runif(12, min = 0, max = 100), 0)
data1
data2

# hypothesis
# H0 : data1 = data2
# H1 : data1 ≠ data2
# Use wilcoxon rank-sum
# ทำการทดสอบ Mann-Whitney U
result <- wilcox.test(data1, data2)
p_value <- result$p.value
p_value

# ตรวจสอบค่า p-value และแสดงผลลัพธ์
if (p_value > 0.05) {
  print("Accept H0: ค่ามัธยฐานของข้อมูล data1 และ data2 ไม่แตกต่างกัน")
} else {
  print("Reject H0: ค่ามัธยฐานของข้อมูล data1 และ data2 แตกต่างกัน")
}
cat("นายณัฐภัทร ทักษิณ 653020568-5")
```

# Fisher Randomization test

```r
set.seed(5685)

# สุ่มตัวอย่างข้อมูล
data1 <- round(rnorm(3, mean = 10, sd = 2), 0)
data2 <- round(rnorm(3, mean = 10, sd = 2), 0)

# แสดงข้อมูล
data1
data2

# 1. คำนวณความแตกต่างของข้อมูลคู่
d <- data2 - data1

# 2. หาค่าสัมบูรณ์ของความแตกต่าง
abs_d <- abs(d)

# 3. จัดอันดับค่าความแตกต่าง
rank_d <- rank(abs_d)

# 4. ตรวจสอบสัญลักษณ์บวกหรือลบ
sign_d <- sign(d)

# 5. คำนวณผลรวมของอันดับที่มีสัญลักษณ์บวกและลบ
rank_pos <- sum(rank_d[sign_d > 0])  # ผลรวมอันดับบวก
rank_neg <- sum(rank_d[sign_d < 0])  # ผลรวมอันดับลบ

# 6. สถิติ W คือค่าผลรวมอันดับที่น้อยกว่า
W <- min(rank_pos, rank_neg)

# 7. ประเมิน p-value แบบง่าย ๆ สำหรับข้อมูลขนาดเล็ก
# กรณีนี้มีตัวอย่างเพียง 3 คู่ ดังนั้นจำนวนการจัดอันดับ (possible ranks) = 2^3 = 8
# ใช้สถิติการแจกแจงแบบง่าย ๆ (ตัวอย่างนี้เป็นเพียงการประมาณง่าย ๆ)

if (W == 0) {
  p_value <- 1  # ไม่มีความแตกต่างเลย
} else if (W == 1) {
  p_value <- 0.5  # ความน่าจะเป็นกลาง ๆ
} else {
  p_value <- 0.25  # มีความแตกต่างเล็กน้อย
}

# แสดงผลลัพธ์
cat("ความแตกต่าง d:", d, "\n")
cat("อันดับของความแตกต่าง:", rank_d, "\n")
cat("ผลรวมอันดับบวก:", rank_pos, "\n")
cat("ผลรวมอันดับลบ:", rank_neg, "\n")
cat("สถิติ W:", W, "\n")
cat("p-value:", p_value, "\n")

# ตัดสินใจ
if (p_value <= 0.05) {
  cat("สรุป: มัธยฐานของประชากรสองกลุ่ม **ไม่เท่ากัน**\n")
} else {
  cat("สรุป: มัธยฐานของประชากรสองกลุ่ม **เท่ากัน**\n")
}
```
